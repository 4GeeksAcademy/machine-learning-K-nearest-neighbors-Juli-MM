# K-nearest neighbors Project 

Movie recommendation system

- Would we be able to predict which movies might or might not be a commercial success? This dataset collects part of the knowledge from the API TMDB, which contains only 5000 movies out of the total number.

#### Step 1: Loading the dataset

```python
import pandas as pd

movies_data = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
movies_data.head()
```

```python
print(movies_data.shape)
print(movies_data.info())
```

```python
credits_data = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')
credits_data.head()
```

```python
print(credits_data.shape)
print(credits_data.info())
```

#### Step 2: Creation of a database

```python
import sqlite3

# Create database in SQLite:

conn = sqlite3.connect('movies_database.db')

# Create and convert the datasets to SQL:

movies_data.to_sql('movies', conn, if_exists='replace', index=False)
credits_data.to_sql('credits', conn, if_exists='replace', index=False)

# Query to create new table:

query = '''
    SELECT *
    FROM movies
    INNER JOIN credits
    USING (title);
'''
# Create new database:

new_data = pd.read_sql_query(query, conn)
conn.close()
```

```python
# Print new dataset to see how it looks like:

new_data.head()
```

```python
# Clean the new dataset:

new_data = new_data[['movie_id','title','overview','genres','keywords','cast','crew']]
new_data.head()
```

```python
print(new_data.shape)
print(new_data.info())
```

```python
# Drop nulls:

#new_data.dropna(inplace = True)
#print(new_data.info())
```

#### Step 3: Transform the data

```python
# Import JSON package:

import json

new_data["genres"] = new_data["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
new_data["keywords"] = new_data["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
new_data["cast"] = new_data["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)
new_data["crew"] = new_data["crew"].apply(lambda x: next((item["name"] for item in json.loads(x) if item["job"] == "Director"), None) if pd.notna(x) else None)
new_data["overview"] = new_data["overview"].apply(lambda x: [x] if pd.notna(x) else None)

new_data.head()
```

```python
# Remove spaces:

def remove_spaces_from_list(lst):
    return [item.replace(" ", "") for item in lst] if isinstance(lst, list) else lst

new_data[["genres", "cast", "keywords"]] = new_data[["genres", "cast", "keywords"]].map(remove_spaces_from_list)
new_data["crew"] = new_data["crew"].str.replace(" ", "")

new_data.head()
```

```python
# Convert the lists to strings in the dataset:

def list_to_string(lst):
    return ",".join(lst) if isinstance(lst, list) else ""

new_data["genres"] = new_data["genres"].apply(list_to_string)
new_data["keywords"] = new_data["keywords"].apply(list_to_string)
new_data["cast"] = new_data["cast"].apply(list_to_string)
new_data["overview"] = new_data["overview"].apply(list_to_string)
```

```python
# Create a single line content in a "Tags" Column:

new_data['overview'] = new_data['overview'].str.replace(",", "")
new_data["tags"] = new_data["overview"] + " " + new_data["genres"] + " " + new_data["keywords"]  + " " + new_data["cast"] + " " + new_data["crew"]
new_data["tags"] = new_data["tags"].str.replace(",", " ")

new_data["tags"][0]
```

#### Step 4: Build a KNN

```python
from sklearn.feature_extraction.text import CountVectorizer

new_data["tags"] = new_data["tags"].fillna("")
cv = CountVectorizer(max_features=5000 ,stop_words='english')
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_lower = movie.lower()  # Convert the input to lowercase
    movie_index = new_data[new_data["title"].str.lower() == movie_lower].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"Similar movies to '{movie}':")
    for i in movie_list:
        print(new_data.iloc[i[0]].title)

user_input = input("Enter a movie name: ")
recommend(user_input)
```
